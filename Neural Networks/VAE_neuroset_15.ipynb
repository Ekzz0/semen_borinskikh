```python

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow import keras
import keras.backend as K
from tensorflow.keras.layers import Dense, Flatten, Reshape, Input, Lambda, BatchNormalization, Dropout
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# стандартизация входных данных
x_train = x_train / 255
x_test = x_test / 255

x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test  = np.reshape(x_test,  (len(x_test),  28, 28, 1))

```

```python

hidden_dim = 2 # размерность скрытого пространства (вектор С.П будет иметь 2 элемента)
batch_size = 60 # должно быть кратно 60 000

def dropout_and_batch(x): # привязка к слою x батч нормализация, а после к этому всему дропаут
    return Dropout(0.3)(BatchNormalization()(x))

input_img = Input((28, 28, 1))
x = Flatten()(input_img)
x = Dense(256, activation='relu')(x)
x = dropout_and_batch(x)
x = Dense(128, activation='relu')(x)
x = dropout_and_batch(x)

z_mean = Dense(hidden_dim)(x) # для вектора мат. ожидания
z_log_var = Dense(hidden_dim)(x) # для вектора логорифма дисперсии 
```

```python
def noiser(args):
    global z_mean, z_log_var
    z_mean, z_log_var = args
    N = K.random_normal(shape=(batch_size, hidden_dim), mean=0., stddev=1.0) # моделирование тензора нормальный величин нужной размерности
    return K.exp(z_log_var / 2) * N + z_mean # формирование вектора h по формуле, согласно схеме 
    # exp(z_log_var / 2) = dispersion (ср.квадр отклон.)

h = Lambda(noiser, output_shape=(hidden_dim,))([z_mean, z_log_var])  # подключение функции к лямбда слою
```

```python
# определим декодер
input_dec = Input(shape=(hidden_dim,))
d = Dense(128, activation='relu')(input_dec)
d = dropout_and_batch(d)
d = Dense(256, activation='relu')(d)
d = dropout_and_batch(d)
d = Dense(28*28, activation='sigmoid')(d)
decoded = Reshape((28, 28, 1))(d)
```

```python
encoder = keras.Model(input_img, h, name='encoder')
decoder = keras.Model(input_dec, decoded, name='decoder')
vae = keras.Model(input_img, decoder(encoder(input_img)), name="vae")
```

```python
# Функция для вычисления критерия качества
def vae_loss(x, y): # x - входные, у - требуемые выходные
    x = K.reshape(x, shape=(batch_size, 28*28)) # преобрзовываем их в двумерные тензоры
    y = K.reshape(y, shape=(batch_size, 28*28))
    loss = K.sum(K.square(x-y), axis=-1) # сумма квадратов рассолгасований между входами и выходами (получается тензор (1 x batch_size)
    kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1) # axis=-1 - вычисление по последней оси 
    return loss + kl_loss
```

```python
vae.compile(optimizer='adam', loss=vae_loss)
```

```python
vae.fit(x_train, x_train, epochs=5, batch_size=batch_size, shuffle=True)
```
